###**Merge e Join**

diferença principal entre o **merge e o join** em pandas está na forma como eles lidam com as colunas de junção.

**merge:** É um método mais genérico e versátil. Você pode especificar colunas diferentes em ambos os dataframes para a junção, utilizando os parâmetros left_on e right_on (ou left_index e right_index para índices).

Exemplo:

```
merged_df = df1.merge(df2, left_on='coluna_df1', right_on='coluna_df2',
how='inner')

```
**join:** É um método mais simples que se baseia principalmente em índices para a junção. Ele junta dois dataframes com base em seus índices.


```
joined_df = df1.join(df2, how='inner')
```

Em resumo, o merge oferece mais flexibilidade ao permitir que você especifique colunas diferentes para a junção, enquanto o join é mais direto, usando principalmente os índices.
